###**Clustering for Anomaly Detection**


**Clustering Techniques**
- Clustering techniques like K-means, DBSCAN, and hierarchical clustering are effective for anomaly detection because they group similar data points and identify outliers based on their distance from cluster centroids.

**K-Means Clustering**
- K-means clustering partitions data into k clusters. - Anomalies are data points far from any cluster center.

```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

# Load dataset from user (replace with actual file path)
file_path = "/content/reduced_dataset.csv"  # Provide the correct path to your dataset
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to check the loaded data
print("Dataset preview:")
print(data.head())

# List the available columns to identify which to use for clustering
print("\nColumns in the dataset:")
print(data.columns)

# Specify the feature columns for clustering (update these based on your dataset)
feature_columns = input("Enter the column names to use for clustering, separated by commas (e.g., 'col1,col2'): ").split(',')

# Check if the specified columns exist in the dataset
for col in feature_columns:
    if col not in data.columns:
        raise ValueError(f"The dataset does not contain the column: {col}")

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(data[feature_columns])

# Anomaly detection based on distance from the cluster centroid
data['distance_from_centroid'] = np.linalg.norm(
    data[feature_columns].values - kmeans.cluster_centers_[data['cluster']],
    axis=1
)

# Set an appropriate threshold for anomaly detection (adjust this as needed)
threshold = float(input("Enter the threshold for anomaly detection (e.g., 0.5): "))

# Identify anomalies based on the threshold
anomalies = data[data['distance_from_centroid'] > threshold]

# Output anomalies
print(f"\nDetected anomalies (distance from centroid > {threshold}):")
print(anomalies)

# Optionally, save the anomalies to a new CSV file
save_file = input("Do you want to save the anomalies to a CSV file? (yes/no): ")
if save_file.lower() == 'yes':
    output_file = input("Enter the output file name (e.g., anomalies.csv): ")
    anomalies.to_csv(output_file, index=False)
    print(f"Anomalies have been saved to {output_file}")
```

##**Dimensionality Reduction**

- Dimensionality reduction is an unsupervised learning technique used to reduce the number of input features in a dataset while retaining its essential structure and information. This is particularly important when working with high-dimensional datasets where a large number of features can lead to problems like increased computational cost, overfitting, or difficulty in visualizing the data. Dimensionality reduction can be broadly categorized into feature selection (selecting the most important features) and feature extraction (transforming features into a lower-dimensional space).




###**1. Principal Component Analysis (PCA)**


- **Purpose:** To transform high-dimensional data into a lower-dimensional space while capturing the maximum variance.

**How it works:**
- PCA identifies the directions (principal components) along which the variance in the data is maximized.
These principal components are orthogonal (independent) linear combinations of the original features.
- The data is then projected onto these components, and only the most significant components (those explaining the most variance) are retained.

**Applications:**
- **Data compression:** Reducing storage space by keeping only significant components.
- **Noise reduction:** By discarding minor components, noise can be filtered out.
- **Preprocessing for supervised learning:** Reducing feature dimensions before applying models.

**Advantages:**
- Computationally efficient and easy to implement.
- Works well for linearly separable data.

**Disadvantages:**
- PCA assumes linear relationships and may not work well with non-linear data.
- The reduced dimensions may lack interpretability.
  - **Use Case Example:** Reducing the dimensionality of images in facial recognition systems while retaining distinguishing features.



```python
# Import required libraries
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/reduced_dataset.csv'
data = pd.read_csv(file_path)

# Select numerical columns for PCA
numerical_columns = ['Inventory', 'Demand', 'Price', 'Sales']
features = data[numerical_columns]

# Step 1: Standardize the numerical data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 2: Apply PCA
n_components = 2  # Number of principal components
pca = PCA(n_components=n_components)
pca_result = pca.fit_transform(scaled_features)

# Step 3: Visualize the PCA results
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c='blue', alpha=0.7)
plt.title(f'PCA Results (First {n_components} Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()

# Step 4: Print the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio for each component:")
for i, var_ratio in enumerate(explained_variance_ratio, start=1):
    print(f"Component {i}: {var_ratio:.4f}")
```

###**2. t-SNE (t-Distributed Stochastic Neighbor Embedding)**


- **Purpose:** A non-linear dimensionality reduction technique primarily used for visualizing high-dimensional data in 2D or 3D space.

**How it works:**
- t-SNE minimizes the divergence between two distributions: one that measures pairwise similarities of the data points in the high-dimensional space and another in the low-dimensional space.
- It focuses on preserving local structures, meaning that similar data points in high-dimensional space remain close in the lower-dimensional visualization.

**Applications:**
- Exploratory data analysis: Visualizing clusters and patterns in data.
- **Anomaly detection:** Identifying outliers in high-dimensional datasets.

**Advantages:**
- Captures non-linear relationships between features.
- Produces highly interpretable visualizations.

**Disadvantages:**
- Computationally expensive, especially for large datasets.
- The results are not deterministic (results can vary for different runs).
  - **Use Case Example:** Visualizing word embeddings (e.g., Word2Vec or GloVe) to analyze relationships between words.



```python
# Import required libraries
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/reduced_dataset.csv'
data = pd.read_csv(file_path)

# Select numerical columns for t-SNE
numerical_columns = ['Inventory', 'Demand', 'Price', 'Sales']
features = data[numerical_columns]

# Step 1: Standardize the numerical data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 2: Apply t-SNE
# Set perplexity to a value smaller than the number of samples
tsne = TSNE(n_components=2, random_state=42, perplexity=2, n_iter=1000)
tsne_result = tsne.fit_transform(scaled_features)

# Step 3: Visualize the t-SNE results
plt.figure(figsize=(8, 6))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c='blue', alpha=0.7)
plt.title('t-SNE Results')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.grid()
plt.show()
```

###**3. Autoencoders**


- **Purpose:** A neural network-based technique used for encoding high-dimensional data into a compressed representation and reconstructing it back to its original form.

**How it works:**
- An autoencoder consists of two parts:
  - **Encoder:** Compresses the input data into a lower-dimensional latent space.
  - **Decoder:** Reconstructs the original data from the compressed representation.

- The network is trained to minimize the reconstruction error between the input and the output.

**Applications:**
- **Feature extraction:** Learning efficient representations of high-dimensional data.
- **Anomaly detection:** Identifying instances with high reconstruction error.
- **Data denoising:** Removing noise from data while retaining the core structure.

**Advantages:**
- Can model complex non-linear relationships.
- Highly flexible and can incorporate additional domain knowledge through neural network architecture.

**Disadvantages:**
- Requires more computational power and larger datasets to train effectively.
- The learned latent space may not always be interpretable.
  - **Use Case Example:** Compressing high-dimensional time-series data for use in predictive maintenance systems.



```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/reduced_dataset.csv'
data = pd.read_csv(file_path)

# Step 1: Select numerical features for Autoencoder
numerical_columns = ['Inventory', 'Demand', 'Price', 'Sales']
features = data[numerical_columns]

# Step 2: Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 3: Split the data into training and testing sets
X_train, X_test = train_test_split(scaled_features, test_size=0.2, random_state=42)

# Step 4: Define the Autoencoder architecture
input_dim = X_train.shape[1]  # Number of features
encoding_dim = 2  # Dimension of the latent space (bottleneck)

# Input layer
input_layer = Input(shape=(input_dim,))
# Encoder layer
encoded = Dense(encoding_dim, activation='relu')(input_layer)
# Decoder layer
decoded = Dense(input_dim, activation='sigmoid')(encoded)

# Autoencoder model
autoencoder = Model(inputs=input_layer, outputs=decoded)

# Compile the model
autoencoder.compile(optimizer='adam', loss='mse')

# Step 5: Train the Autoencoder
history = autoencoder.fit(
    X_train, X_train,
    epochs=50,
    batch_size=2,
    validation_data=(X_test, X_test),
    verbose=1
)


# Step 6: Evaluate reconstruction on test data
reconstructed = autoencoder.predict(X_test)

# Visualize the original vs reconstructed data
plt.figure(figsize=(12, 6))
plt.plot(X_test[0], label='Original', marker='o')
plt.plot(reconstructed[0], label='Reconstructed', marker='x')
plt.title("Original vs Reconstructed Data (Sample)")
plt.legend()
plt.grid()
plt.show()

# Step 7: Visualize training and validation loss
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Autoencoder Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()
```

###**4. UMAP (Uniform Manifold Approximation and Projection)**


- **Purpose:** A non-linear dimensionality reduction technique that preserves both local and global structures of data better than t-SNE.

**How it works:**
- UMAP constructs a graph of data points in high-dimensional space based on their local neighborhood relationships.
- It then optimizes a low-dimensional representation that maintains these relationships.
- Unlike t-SNE, UMAP emphasizes both local (similar data points) and global (large-scale) structures.

**Applications:**
- Visualizing high-dimensional data with meaningful clusters.
- Preprocessing step for other machine learning tasks.
Faster and more accurate embedding of data compared to t-SNE.

**Advantages:**
- Faster than t-SNE, especially for large datasets.
- Better preservation of both local and global structures.
- Deterministic results (does not vary with each run, unlike t-SNE).

**Disadvantages:**
- Requires tuning of hyperparameters like the number of neighbors and minimum distance.
- Results can vary significantly based on the dataset and parameter settings.

##**Anomaly Detection**

- Anomaly Detection is a technique used in unsupervised learning to identify patterns in data that do not conform to expected behavior. These patterns are often referred to as "outliers" or "novelty," and detecting them can be crucial in various applications such as fraud detection, network security, health monitoring, and quality control.

**Anomalies can be classified into:**
- **Point Anomalies:** A single data point is different from the rest of the data.
- **Contextual Anomalies:** An observation that might be normal in some context but is an anomaly in another context (e.g., a high temperature at night).
- **Collective Anomalies:** A collection of data points that are anomalous when considered together but not individually (e.g., unusual traffic patterns over time).

Unsupervised learning methods for anomaly detection are ideal because they don’t require labeled data. The key focus of such methods is to identify instances that deviate significantly from the general data distribution.

**Common Approaches to Anomaly Detection:**
- **Clustering-based Methods:** These methods assume that normal data points form dense regions, while anomalies are isolated. K-means clustering or DBSCAN can be used for this purpose.
- **Density-based Methods:** Methods like Local Outlier Factor (LOF) and Isolation Forest work by estimating the density of data points in a local region and flagging low-density regions as anomalies.
- **Distance-based Methods:** Anomalies are detected by looking at the distance between a point and its neighbors. One of the most common distance-based methods is the K-Nearest Neighbors (KNN).
- **Model-based Methods:** These involve fitting a model to the data and flagging points that deviate from the model as anomalies. Examples include Gaussian Mixture Models (GMM) and One-Class SVM.
- **Statistical Methods:** Statistical tests such as Z-score or Chi-square test can also be used to identify outliers based on the distribution of data.



###**1. K-Means Clustering-based Anomaly Detection**


- **Explanation:** K-means is used to partition the data into clusters, and anomalies are detected based on how far they are from the centroids of these clusters.

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Load dataset
file_path = input("/content/smart_supply_chain_sample.csv")
data = pd.read_csv(file_path)

# Select features for clustering (adjust column names based on your dataset)
feature_columns = input("Enter the feature columns for clustering, separated by commas (e.g., 'feature1,feature2'): ").split(',')
data_selected = data[feature_columns]

# K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(data_selected)

# Compute the distance from the centroid
data['distance_from_centroid'] = np.linalg.norm(data_selected.values - kmeans.cluster_centers_[data['cluster']], axis=1)

# Set a threshold for anomaly detection
threshold = float(input("Enter the distance threshold for anomaly detection (e.g., 2.5): "))

# Flag anomalies
anomalies = data[data['distance_from_centroid'] > threshold]

# Display anomalies
print(f"Detected anomalies: {anomalies}")

# Optionally, save anomalies to a CSV
save_option = input("Do you want to save the anomalies to a CSV file? (yes/no): ")
if save_option.lower() == 'yes':
    anomalies.to_csv("anomalies.csv", index=False)
    print("Anomalies saved to 'anomalies.csv'")
```

##**Time Series Forecasting**

- Time series forecasting involves predicting future values based on previously observed values. It has wide applications in various domains like finance, healthcare, weather forecasting, sales prediction, and inventory management.

**Key Characteristics of Time Series Data:**
- **Trend:** A long-term movement or direction in data (e.g., increasing or decreasing over time).
Seasonality: Patterns that repeat at regular intervals (e.g., daily, weekly, or yearly).
- **Noise:** Random fluctuations that are difficult to predict.
- **Cyclic Patterns:** Long-term fluctuations that do not have a fixed period.

**Steps in Time Series Forecasting:**
- **Data Preprocessing:** This step involves cleaning the data by handling missing values, smoothing out noise, and making the data stationary (removing trends).
- **Exploratory Data Analysis (EDA):** Visualize trends, seasonality, and anomalies in the data using plots and statistical techniques.
- **Model Building:** ARIMA, among other models, is used to build a predictive model based on past data.
- **Model Evaluation:** Evaluate the model using performance metrics such as Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and others.
- **Forecasting:** Use the trained model to predict future values.


###**ARIMA Model (AutoRegressive Integrated Moving Average)**

**Components of ARIMA:**
- **AutoRegressive (AR):** This part of the model regresses the time series against itself, using previous observations. The AR part is defined by a parameter p, which represents the number of lagged observations included in the model.
  - For example, if p = 2, the model will use the values from two previous time steps to predict the next value.
- **Integrated (I):** This part involves differencing the time series to make it stationary. The I component is represented by the parameter d, which indicates the number of differencing operations needed.
  - If a time series has a trend, differencing will remove the trend and help make the data stationary. d = 1 means we subtract each observation from the previous one to remove the trend.
- **Moving Average (MA):** The MA component models the relationship between an observation and the residual errors from a moving average model applied to lagged observations. The parameter q represents the number of lagged forecast errors included in the model.



###**Model Parameters:**


- **p (AR order):** The number of lagged terms in the autoregressive model.
- **d (Differencing):** The number of times the series is differenced to make it stationary.
- **q (MA order):** The number of lagged forecast errors in the moving average model.

**Identifying the right values for p, d, and q:**
- ACF (Autocorrelation Function) and PACF (Partial Autocorrelation Function) plots are typically used to determine the initial parameters (p and q).
- ACF helps identify the MA order by showing the correlation between a series and its lags.
- PACF is used to identify the AR order by showing the correlation of the series with its lagged values after removing the influence of earlier lags.


###**Steps in ARIMA Model Application:**


**Step 1: Stationarity Check**
- Before applying ARIMA, check if the time series is stationary. A stationary time series has constant mean, variance, and autocorrelation over time.
- If the series is non-stationary, apply differencing until stationarity is achieved.

**Step 2: Model Selection (p, d, q)**
- Use ACF and PACF plots to choose the initial parameters.
- Grid search can also be used to find the optimal combination of p, d, and q using cross-validation.

**Step 3: Fit the ARIMA Model**
- Fit the ARIMA model to the data using the statsmodels library or other packages such as pmdarima.


**Example code to fit an ARIMA model:**

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load and preprocess data (if not already done)
file_path = '/content/reduced_dataset.csv'
dataset = pd.read_csv(file_path)
dataset['Date'] = pd.to_datetime(dataset['Date'])
time_series_data = dataset.set_index('Date')
sales_time_series = time_series_data['Sales'].resample('D').sum()

# Plot the time series data
plt.figure(figsize=(10, 6))
sales_time_series.plot()
plt.title('Daily Sales Time Series')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()

# Fit an ARIMA model
# (p, d, q) are the parameters of the ARIMA model; modify based on your analysis
model = ARIMA(sales_time_series, order=(1, 1, 1))
arima_result = model.fit()

# Display the summary of the ARIMA model
print(arima_result.summary())

# Plot the model's diagnostics
arima_result.plot_diagnostics(figsize=(10, 8))
plt.show()

# Forecasting future values
forecast_steps = 30  # Number of days to forecast
forecast = arima_result.forecast(steps=forecast_steps)

# Plot the forecast
plt.figure(figsize=(8, 4))
sales_time_series.plot(label='Observed', legend=True)
forecast.plot(label='Forecast', legend=True)
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

##**Recommender System for Restocking**

A Recommender System is an information filtering system designed to predict the preferences or interests of a user and provide recommendations accordingly. These systems are widely used in e-commerce, entertainment, social media, and various online platforms to enhance user experiences by offering personalized suggestions.

**Objectives of Recommender Systems:**
- **Personalization:** Tailor the content for individual users.
- **Increased Engagement:** Improve user interaction with the platform.
- **Revenue Boost:** Drive sales or subscriptions through recommendations.
- **Decision Support:** Help users make informed choices.
- **Customer Retention:** Enhance user satisfaction to retain customers.



**Key Concepts in Recommender Systems**

**Cold Start Problem:**
- Occurs when insufficient data about users or items hinders recommendations.
- Solutions include asking users to complete preference surveys or using hybrid approaches.

**Sparsity:**
- Most recommendation datasets are sparse (users interact with only a small portion of items).
- Techniques like matrix factorization or deep learning can help mitigate this.

**Scalability:**
- Handling recommendations efficiently for millions of users and items is crucial.
- Distributed computing and optimized algorithms address this challenge.

**Diversity and Serendipity:**
- Ensures recommendations are not monotonous or overly predictable.
- Introduces novel or unexpected items to users.


###**Types of Filtering**

**Content-Based Filtering:**
- Recommendations are based on the features of items and a user’s past interactions.
- Uses item descriptions and user profiles to match preferences.
  - Example: A movie platform recommending films similar to those a user has rated highly.

**Advantages:**
- Personalized.
- Does not require data from other users.

**Disadvantages:**
- Limited by the features of the content.
- Struggles to recommend diverse items (serendipity problem).


**Collaborative Filtering:**
- Recommendations are based on the behavior and preferences of other users.

**Two main types:**
- **User-Based Collaborative Filtering:** Finds users with similar tastes.
- **Item-Based Collaborative Filtering:** Finds items that are commonly liked by users with similar preferences.

**Advantages:**
- Works well without detailed item information.
- Can uncover hidden relationships between items.

**Disadvantages:**
- Cold start problem (difficulty recommending to new users/items).
- Scalability issues with a large user or item base.

**Hybrid Recommender Systems:**
- Combines multiple recommendation techniques to overcome the limitations of individual methods.

**Examples:**
 - Combining content-based and collaborative filtering.
 - Adding business rules or contextual information.

**Advantages:**
- Improved accuracy and diversity.
- Balances strengths and weaknesses of individual systems.

**Knowledge-Based Recommender Systems:**
- Recommendations are based on explicit user requirements or domain knowledge.
- Used in domains where collaborative or content-based methods are unsuitable (e.g., luxury items, rare purchases).


###**Code Prompts to Illustrate**

- **Content-Based Recommender System**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
file_path = '/content/reduced_dataset.csv'
dataset = pd.read_csv(file_path)

# Prepare product-based features for recommendation (e.g., based on 'Category')
dataset['Category'] = dataset['Category'].fillna('')
dataset['Product'] = dataset['Product'].fillna('')

# Use TF-IDF vectorizer to convert product categories into feature vectors
tfidf = TfidfVectorizer(stop_words='english')
category_matrix = tfidf.fit_transform(dataset['Category'])

# Compute similarity between products based on their categories
cosine_sim = cosine_similarity(category_matrix, category_matrix)

# Function to recommend products based on a product ID
def recommend_products(product_id, top_n=5):
    idx = dataset.index[dataset['Product'] == product_id].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    product_indices = [i[0] for i in sim_scores]

    return dataset.iloc[product_indices][['Product', 'Category', 'Price']]

# Example: Recommend similar products to 'Product_360'
recommended_products = recommend_products('Product_360', top_n=5)
print(recommended_products)
```

- **Collaborative Filtering (Matrix Factorization) Recommender System**

```python
import pandas as pd
from sklearn.decomposition import NMF
import numpy as np

# Load the dataset
file_path = '/mnt/data/reduced_dataset.csv'
dataset = pd.read_csv(file_path)

# Create a user-product matrix based on Sales or Demand (pivot the dataset)
user_product_matrix = dataset.pivot_table(index='Warehouse', columns='Product', values='Sales', aggfunc='sum', fill_value=0)

# Perform NMF (Non-negative Matrix Factorization) to factorize the matrix
model = NMF(n_components=5, random_state=42)
W = model.fit_transform(user_product_matrix)  # User matrix
H = model.components_  # Item matrix

# Reconstruct the matrix and predict sales for all user-product pairs
predicted_sales = np.dot(W, H)

# Convert the predicted sales back to a DataFrame
predicted_df = pd.DataFrame(predicted_sales, columns=user_product_matrix.columns, index=user_product_matrix.index)

# Function to recommend top products for a specific warehouse
def recommend_products_for_warehouse(warehouse, top_n=5):
    warehouse_predictions = predicted_df.loc[warehouse].sort_values(ascending=False)
    top_products = warehouse_predictions.head(top_n)
    return top_products

# Example: Recommend top 5 products for 'Warehouse_11'
recommended_products_warehouse = recommend_products_for_warehouse('Warehouse_11', top_n=5)
print(recommended_products_warehouse)
```

**Key Points:**
- Content-Based Filtering uses product characteristics like "Category" or "Price" to recommend similar products based on the similarity of features.
- Collaborative Filtering (Matrix Factorization) uses interactions (e.g., sales or demand) to recommend products based on user-product interactions.
